```python
# Create template for prompt
# Scenario 1
prompt_one = "கீழே கொடுக்கப்பட்டுள்ள வார்த்தைகளையும் தலைப்பையும் பயன்படுத்தி சிறு கதை எழுதுக.\nவார்த்தைகள்: {}\nதலைப்பு: {}\n"

# Scenario 2
prompt_two = "கீழே கொடுக்கப்பட்டுள்ள கதைக்குப் பொருத்தமான தலைப்பைக் கொடு.\nகதை: {}\n"
completion_two = "கொடுக்கப்பட்டுள்ள கதைக்குப் பொருத்தமான தலைப்பு '{}' என்பதாகும்."

story_data = pd.DataFrame(columns=["inputs", "targets"])
story_title = pd.DataFrame(columns=["inputs", "targets"])
```

```python
# Read the json files and combine
#filenames = next(os.walk('.'))[2] #use this once all the stories are completed
#print(filenames)
filenames = ['zen_stories.json',
             'mariyadai_raman_stories.json',
             'aesop_fables_stories.json',
             'siruvarmalar.json',
             'oshos_stories.json',
             'mulla_stories.json',
             'akbar_birbal_stories.json',
             'brahmartha_guru_stories.json',
             'multy_taste_stories.json',
             'panchatantra_stories.json',
             'short_stories.json',
             'thenali_raman_stories.json'
             ]


prompt_list = []
story_list = []
title_list = []

for file in filenames:
  with open(file) as f:
    data = ast.literal_eval(str(json.load(f)))

  for entry in data:
    # Fix issues with the story
    story = entry['story'].replace("`", "'")
    story = story.replace("’", "'")
    story = story.replace("’ ", "'")
    story = story.replace(" ’", "'")
    story = story.replace("“ ", '"')
    story = story.replace(" ”", '"')
    story = story.replace("“", '"')
    story = story.replace("”", '"')
    story = story.replace("\'\'", '"')
    story = story.replace("\'", '"')
    tory = story.replace("\'\' ", '"')
    story = story.replace("\' ", '"')
    story = story.replace('\u200c', '')
    story = story.replace("'' ''", '", "')
    story = story.replace('" "', '", "')
    story = story.replace("'' ", '"')
    story = story.replace(" ''", '"')

    # Fix issues with title
    title =  entry['story_title'].replace(".", '')

    # Add prompt and completion for scenario 1
    prompt_list.append(prompt_one.format(entry['characters'], entry['story_title']))
    story_list.append(story)

    # Add prompt and completion for scenario 2
    prompt_list.append(prompt_two.format(entry['story']))
    story_list.append(completion_two.format(entry['story_title']))

story_data['inputs'] = prompt_list
story_data['targets'] = story_list
```
